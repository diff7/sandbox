```python
def plot_spectr(signal, axes=None):
    mag, _ = librosa.magphase(librosa.stft(signal, 
                         n_fft=320, 
                         hop_length=160, 
                         win_length=320))
    librosa.display.specshow(
                    librosa.amplitude_to_db(mag),
                    cmap="jet",
                    y_axis="linear",
                    ax=axes,
                    sr=1600,
                )
```
