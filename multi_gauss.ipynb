```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
means = np.random.normal(size=2)
# example cov matrix
cov = np.array([[7,2],[2,1]])
random_gauss = np.random.multivariate_normal(means, cov, size = 3000)
```

```python
k = 0 
for i in range(10):
    k +=100 
    plt.scatter(random_gauss[0:k,0], random_gauss[0:k,1])
    plt.show()
```
